```python
# Need to make the column names the same as the dataframe col names

try:
    # Connect to PostgreSQL
    conn = psycopg2.connect(
    dbname='brdb',
    user='brandon',
    password='',
    host='localhost',
    port='5432'
    )

    cur = conn.cursor()

    # Read CSV file into a pandas DataFrame
    df = pd.read_csv(csv_folder)

    df = df.rename(columns={'RANK': 'Rank', 'MOVEMENT': 'Movement', 
                            'PLAYER_ID': 'PlayerID', 'PLAYER': 'Player', 'TOTAL SG:T': 'Total_SG_T'
                            , 'TOTAL SG:T2G': 'Total_SG_T2G', 'TOTAL SG:P':'Total_SG_P'
                            , 'MEASURED ROUNDS':'Measured_Rounds' })
    print(df.columns)

    columns_to_clean = ['par', 'Week_Of_Season', 'Year','Rank','PlayerID','Measured_Rounds']
    for col in columns_to_clean:
        df[col] = df[col].apply(lambda x: None if pd.isna(x) else int(x))

  

    # Insert data into the existing table
    for _, row in df.iterrows():
        insert_query = f"""
        INSERT INTO SG_Data ({', '.join(df.columns)})
        VALUES ({', '.join(['%s' for _ in df.columns])});
        """
        #cur.execute(insert_query, tuple(row))
        try:
            cur.execute(insert_query, tuple(row))
        except Exception as e:
            print(f"Error occurred for row {tuple(row)}: {e}")


    # Commit changes
    conn.commit()
    print(f"Data from {csv_folder} has been uploaded successfully!")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Close the connection
    if cur:
        cur.close()
    if conn:
        conn.close()

# end of script 
```
