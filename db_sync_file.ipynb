```python
import subprocess
import os

NEON_URL = os.getenv("NEON_DATABASE_URL")
```

```python
def sync_local_to_neon(db_name="pga_stats"):
    """Push local changes to Neon"""
    dump_file = f"/tmp/{db_name}_sync.sql"
    subprocess.run(["pg_dump", "-h", "localhost", "-U", "your_user", 
                     "-d", db_name, "-F", "p", "-f", dump_file])
    subprocess.run(["psql", NEON_URL, "-f", dump_file])
    print(f"Synced {db_name} to Neon")
```
